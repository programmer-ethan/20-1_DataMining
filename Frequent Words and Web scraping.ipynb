### 1.2 Plot word frequency (5 pts)

- Sort the word frequency in descending order
- Plot the word frequency
- Plot the word frequency in log-log plot.

````python
#help(plt.plot)
````

````python
# YOUR CODE MUST BE HERE
import matplotlib.pyplot as plt
import re, requests,math
from collections import Counter
import numpy as np
url="https://www.gutenberg.org/files/2701/old/moby10b.txt"
mobydick=requests.get(url).text

words=re.split('[^a-zA-Z]+',mobydick)
counter= Counter(word.lower()
                for word in words
                if word)

words_sorted=counter.most_common()

word_dict=dict(words_sorted)
word_counts=list(word_dict.values()) 
xs=[i+1 for i,_ in enumerate(word_counts)]
log_xs=[math.log10(x) for x in xs]
log_counts=[math.log10(x) for x in word_counts]

plt.plot(xs,word_counts)
plt.xlabel("words")
plt.ylabel("frequency")
plt.title("words frequency")
plt.show()
plt.plot(log_xs,log_counts)

plt.xlabel("log words")
plt.ylabel("log frequency")
plt.title("words frequency: log-log plot")
plt.show()
````

````python
# YOUR CODE MUST BE HERE
import re, requests
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
url="https://www.gutenberg.org/files/2701/old/moby10b.txt"
mobydick=requests.get(url).text
idx=0

words=re.split('[^a-zA-Z]+',mobydick)
counter= Counter(word.lower()
                for word in words
                if word)
words_sorted=counter.most_common()

object_words=[]
for key,count in words_sorted:
    if key not in stopwords:
        object_words.append((key,count))
        idx+=1
    if idx==10:
        break
print(object_words)
word_dict=dict(object_words)
wordcloud= WordCloud().generate_from_frequencies(word_dict)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
        
````

````python
# YOUR CODE MUST BE HERE
from bs4 import BeautifulSoup
import requests
from collections import defaultdict
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

major_init=['Architecture', 'Artificial Intelligence', 'Computational Biology', 'Computer Graphics', 'Computer Security', 'Computer Systems', 'Computer Vision', 'Data Science', 'Education', 'Human-Computer Interaction (HCI)', 'Machine Learning', 'Natural Language Processing', 'Networking', 'Operating/Distributed Systems', 'Programming Systems and Verification', 'Robotics', 'Theory']
xs=['ARC', 'AI', 'BIO', 'CG', 'SEC', 'SYS', 'VIS', 'DS', 'EDU', 'HCI', 'ML', 'NLP', 'NET', 'ODS', 'PS', 'ROB', 'TH']
url = "https://cs.stanford.edu/research?items_per_page=All&field_faculty_status_value=active"
soup = BeautifulSoup(requests.get(url).text, 'html5lib')

quarter=0
majors=[]
assistant_values=[]
associate_values=[]
full_professors_values=[]

assistant_counts=defaultdict(float)
associate_counts=defaultdict(float)
full_professors_counts=defaultdict(float)

for major in major_init:
    assistant_counts[major]+=0
    associate_counts[major]+=0
    full_professors_counts[major]+=0

for i,td in enumerate(soup.tbody('td')):
    if quarter==2:
        prof=td.text.strip()
    if quarter==3:
        text=td.prettify()
        words=re.split('<.*>\n',text)
        
        for word in words:
            word=word.strip()
            if word!="":
                majors.append(word)
        n=len(majors)
        
        for major in majors:
            if prof=='Assistant Professor':
                assistant_counts[major]+=1/n
            if prof=='Associate Professor':
                associate_counts[major]+=1/n
            if prof=='Professor':
                full_professors_counts[major]+=1/n        
        
        quarter=-1
        majors=[]
    quarter+=1
assistant_counts=sorted(assistant_counts.items())
associate_counts=sorted(associate_counts.items())
full_professors_counts=sorted(full_professors_counts.items())

for _,y in assistant_counts:
    assistant_values.append(y)
for _,y in associate_counts:
    associate_values.append(y)
for _,y in full_professors_counts:
    full_professors_values.append(y)
    
xs_num=[i+0.1 for i,_ in enumerate(xs)]
np_assistant_values=np.array(assistant_values)
np_associate_values=np.array(associate_values)
np_full_professors_values=np.array(full_professors_values)
bars = np.add(np_assistant_values, np_associate_values).tolist()

plt.figure(figsize=(17, 5))
p1=plt.bar(xs_num,np_assistant_values)
p2=plt.bar(xs_num,np_associate_values, bottom=np_assistant_values)
p3=plt.bar(xs_num,np_full_professors_values, bottom=bars)
plt.legend((p1[0], p2[0], p3[0]), ('assistant professor', 'associate professor','professor'),loc=9)
plt.title("Stanford Univ. faculty research areas")
plt.xlabel("Research Areas")
plt.ylabel("# of faculty members")
plt.xticks([i+0.1 for i,_ in enumerate(xs)],xs)
plt.show
````
